```python
df_copy=df
def bar_chart(colunm):
    survived=df_copy[df_copy["Survived"]==1][colunm].value_counts()
    dead=df_copy[df_copy["Survived"]==0][colunm].value_counts()
    
    df_new=pd.DataFrame([survived,dead])
    df_new.index=["Survived","Dead"]
    df_new.plot(kind="bar",stacked=True,figsize=(10,5))
```
